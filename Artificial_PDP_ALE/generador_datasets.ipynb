# SENO

```python
import numpy as np
import pandas as pd

# Definir la función objetivo con un pequeño error aleatorio
def target_function(x1, x2, x3):
    y = np.abs(np.sin(2 * np.pi * x1 - x2 + 2 * np.pi * (x3)**2))
    # Agregar un pequeño error aleatorio
    error = np.random.normal(loc=0, scale=0.05)
    if y + error<=0.5:
        return 0
    return 1

# Generar valores aleatorios para las variables entre 0 y 1
num_samples = 5000
x1_values = np.random.rand(num_samples)
x2_values = np.random.rand(num_samples)
x3_values = np.random.rand(num_samples)

# Calcular los valores de la función para cada conjunto de variables
y_values = [target_function(x1, x2, x3) for x1, x2, x3 in zip(x1_values, x2_values, x3_values)]

# Crear un DataFrame de pandas con los valores
data = pd.DataFrame({'X1': x1_values, 'X2': x2_values, 'X3': x3_values, 'Y': y_values})

# Guardar los datos en un archivo CSV
data.to_csv('seno.csv', index=False)
```

# Superficie

```python
import numpy as np
import pandas as pd

# Definir la función objetivo con un pequeño error aleatorio
def target_function(x1, x2):
    y = - 2 * (x1)**2 - 2 * (x2)**2 + 2 * x1 + 2 * x2
    # Agregar un pequeño error aleatorio
    error = np.random.normal(loc=0, scale=0.0)
    if y + error<=0.5:
        return 0
    return 1

# Generar valores aleatorios para las variables entre 0 y 1
num_samples = 5000
x1_values = np.random.rand(num_samples)
x2_values = np.random.rand(num_samples)

# Calcular los valores de la función para cada conjunto de variables
y_values = [target_function(x1, x2) for x1, x2 in zip(x1_values, x2_values)]

# Crear un DataFrame de pandas con los valores
data = pd.DataFrame({'X1': x1_values, 'X2': x2_values, 'Y': y_values})

# Guardar los datos en un archivo CSV
data.to_csv('superficie_0.csv', index=False)
```
